```python
# Import dependencies
import pandas as pd

# Read CSV file for 2021
HAPS2021=pd.read_csv("daily_HAPS_2021.csv")

HAPS2021
```

```python
# Read CSV file for 2020
HAPS2020=pd.read_csv("daily_HAPS_2020.csv")

HAPS2020
```

```python
# Join the datasets into one data frame
combined_2020_2021_df=pd.concat([HAPS2020, HAPS2021])
print(combined_2020_2021_df)
```

```python
# Filter out by the Parameters: Benzene, Formaldehyde, 1,3 Butadiene and Lead PM2.5 LC
parameter_df=combined_2020_2021_df[(combined_2020_2021_df["Parameter Name"]=="Benzene") | (combined_2020_2021_df["Parameter Name"]=="Formaldehyde") | (combined_2020_2021_df["Parameter Name"]=="1,3-Butadiene") | (combined_2020_2021_df["Parameter Name"]=="Lead PM2.5 LC")]
parameter_df["Parameter Name"].unique()
```
